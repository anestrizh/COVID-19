Immun_memory2, все из списка Acute_illness2, у кого тест на имунную память позже положителен, чем на острую инфекцию

```python
df = pd.read_csv('/content/drive/MyDrive/Диплом/data_for_science (1).csv', sep=';')
```

```python
df = df[df['idx'].isin(Acute_illness2)]
```

```python
df = df[(df['mkb10_code'] == 'U07.2')]
```

```python
df
```

```python
Immun_memory2 = pd.DataFrame(Acute_illness2, columns=['idx'])
```

```python
def ImmGM ( idx ):
  return min(np.hstack([df[(df['idx'] == idx) & (df['IgM'] < 1) & (df['IgG'] > 10)]['get_date_at'], '2050-01-01']))
```

```python
def ImmM ( idx ):
  return min(np.hstack([df[(df['idx'] == idx) & (df['IgM'] > 1)]['get_date_at'], '2050-01-01']))
```

```python
#np.vectorize(ImmM)(Immun_memory2['idx'])
```

```python
#Когда тест IgM был > 1
Immun_memory2['test_day_M'] = np.vectorize(ImmM)(Immun_memory2['idx'])
```

```python
#Когда тест IgM < 1 & IgG > 10
Immun_memory2['test_day_G_M'] = np.vectorize(ImmGM)(Immun_memory2['idx'])
```

```python
Immun_memory2 = Immun_memory2[Immun_memory2['test_day_G_M'] != '2050-01-01']
```

```python
Immun_memory2 = Immun_memory2[Immun_memory2['test_day_G_M'] > Immun_memory2['test_day_M']]
Immun_memory2
```

```python
pd.DataFrame(Immun_memory2['idx'].unique(), columns=['idx']).to_excel('/content/drive/MyDrive/Диплом/Immun_memory2.xlsx')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Диплом/data_for_science (1).csv', sep=';')
```

```python
df = df[(df['mkb10_code'] == 'U07.2')]
```

```python
def ImmM ( idx ):
  return min(df[(df['idx'] == idx) & (df['IgM'] > 1)]['get_date_at'])
```

```python
np.vectorize(ImmM)(Immun_memory2['idx'])
```

```python
No_immun_memory2 = pd.DataFrame(Acute_illness2, columns=['idx'])
```

```python
#Когда тест IgM был > 1
No_immun_memory2['test_day_M'] = np.vectorize(ImmM)(No_immun_memory2['idx'])
```

```python
def ImmGM_no ( idx ):
  return min(np.hstack([df[(df['idx'] == idx) & (df['IgM'] < 1) & (df['IgG'] < 10)]['get_date_at'], '2050-01-01']))
```

```python
np.vectorize(ImmGM_no)(No_immun_memory2['idx'])
```

```python
#Когда тест IgM < 1 & IgG > 10
No_immun_memory2['day_M_G'] = np.vectorize(ImmGM_no)(No_immun_memory2['idx'])
```

```python
No_immun_memory2 = No_immun_memory2[No_immun_memory2['day_M_G'] != '2050-01-01']
```

```python
No_immun_memory2 = No_immun_memory2[No_immun_memory2['day_M_G'] > No_immun_memory2['test_day_M']]
```
